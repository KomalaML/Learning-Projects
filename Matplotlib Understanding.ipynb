Matplotlib plot Understanding

```python

# this display plots in inline when using jupyter notebooks, don't need when you are running on python is 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
plt.plot()
plt.show()
```

```python
plt.plot([1,2,3]) #when you don't give axis matplot automatically takes index as x axis and y axis the array you give
plt.show()
```

```python
x= np.random.randint(0, 10, 5)
y = np.random.randint(1,11,5)
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,6,50)
#x=np.random.randint(0, 10, 5)
x[:30]
fig, ax=plt.subplots() #shortcut to have figure and have axis 
ax.plot(x,x**2) #creates prabola curve

plt.show()
```

```python
fig,ax1=plt.subplots()
ax1.plot(x,x**2,color='green', label='x2')
ax1.set_ylabel('x2',color='blue')
plt.show()

```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.tan(x))
plt.show()
```

```python
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x, bins=[1,10,20,20,40])
plt.show()
```

```python
df=pd.read_csv('heart-disease.csv')
df.head()
```

```python
plt.hist(df['age'],bins=50,color='blue',edgecolor='black')
plt.xlabel('Age')
plt.title('Age distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.scatter(df['age'],df['chol'],c=df['trestbps'],cmap='coolwarm')
plt.xlabel('Age')
plt.title('Age vs chol')
plt.ylabel('Cholestrol')
plt.colorbar(label='trestbps')
plt.show()
```
