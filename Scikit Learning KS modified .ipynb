```python
#Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
heart_disease=pd.read_csv("heart-disease.csv")
heart_disease
```

```python
X = heart_disease.drop("chol",axis=1) # this considers all the data such as age sex cp as a one variable and considered as X
y = heart_disease["chol"] # for Y just chol
```

```python
X
y
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)
clf.get_params() # gives the number of hyperparameters that are present in randomforest classifier

# classify the patients do they have any heart disease 
```

```python
from sklearn.model_selection import train_test_split #function called train_test_split which is dividing a dataset into training and testing subsets
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
clf.fit(X_train,y_train)
```

```python
y_preds=clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
clf.score(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test,y_preds))
```

```python
confusion_matrix(y_test,y_preds)
```

```python
accuracy_score(y_test,y_preds)
```
